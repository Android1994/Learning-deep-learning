```python
from keras.models import load_model
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import ModelCheckpoint
batch_size=16
epochs=20
train_data = pd.read_csv('data/train-true.csv')
num_of_imgs=train_data.shape[0]

# #model = create_inception_v4() ###14944classes
#model = load_model('temp3.h5')
#model.compile(loss = 'categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

train_images_path="data/train_img2/"

checkpoint = ModelCheckpoint(filepath="models/checkpoint_weights.hdf5")#.{epoch:d}
callbacks=[checkpoint]

train_datagen = ImageDataGenerator(
    shear_range=0.2,
    zoom_range=[0.5, 2.0],
    horizontal_flip=True,
channel_shift_range=20.)

train_generator = train_datagen.flow_from_directory(
        train_images_path,  
        target_size=(224, 224),
        color_mode="rgb",
        batch_size=batch_size,
        shuffle=True,
        seed=30)  

#test_datagen = ImageDataGenerator(rescale=1./255)
# validation_generator = test_datagen.flow_from_directory(
#         validate_images_path,
#         target_size=(64, 64),
#         batch_size=batch_size)
# early_stopping = keras.callbacks.EarlyStopping(monitor='val_acc', min_delta=0, patience=3, verbose=1, mode='auto')

steps_per_epoch = int(np.ceil(num_of_imgs / float(batch_size)))

history = model.fit_generator(
        train_generator,
        steps_per_epoch=steps_per_epoch,
        epochs=epochs,
        verbose=1,
        callbacks=callbacks,
        workers=4)
```

```python
#####submission.csv
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
from PIL import Image
import csv

test_data = pd.read_csv('data/test-true.csv')
    
print(test_data.shape[0]) ###12250290-1225028 ##train_data.shape[0]
file_dir="data/test_img/"

for i in range(0,test_data.shape[0]):
    _id=test_data['id'][i]
    #_landId
    
    file_path="data/submission.csv"
    
    img_path=file_dir+_id+".jpg"
    im = Image.open(img_path).convert('RGB') #:1,L,P,RGB,RGBA,CMYK,YCbCr,I,F
    imResize = im.resize((224, 224))
    imageData = np.array(imResize)

    X = imageData.reshape(1,224,224,3).astype('float32')
    
    pred = model.predict(X)
    class_i = np.argmax(pred[0])
    prob="%0.2f"%(pred[0][class_i])
    
    _landId=''
    for (key, value) in train_generator.class_indices.items():
        if value==class_i:
            _landId=key
    
    res=_landId+' '+prob
    
    
    if os.path.exists(img_path):
        if os.path.exists(file_path):
            da = [_id,res]
            # 
            csvFile = open(file_path, "a+")
            writer = csv.writer(csvFile)
            # 
            writer.writerow(da)
            csvFile.close()
        else:
            fileHeader = ["id", "landmark_id"]
            # 
            da = [_id,res]
            # 
            csvFile = open(file_path, "a+")
            writer = csv.writer(csvFile)
            # 
            writer.writerow(fileHeader)
            writer.writerow(da)
            csvFile.close()
```
